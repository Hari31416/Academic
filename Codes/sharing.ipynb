### Google Colab

The link for Colab is the easiest way to share the notebook. It not only allows to view the notebook it also allows you to play with it. The link for this is:
`https://colab.research.google.com/github/<user>/<file-path>?hl=en`

For example, the link for this notebook is:
`https://colab.research.google.com/github/Hari31416/PYL800/blob/main/Codes/sharing.ipynb?hl=en`

The final query parameter is `hl=en` which is the language of the notebook.

Note: In order for this method or any of the other shown to work, the file must be on the github repository.

### Github

Github provides facility of rendering just the notebook in the browser. The link for this is:
`https://github.com/<user>/<file-path>`

For example, the link for this notebook is:
`https://github.com/Hari31416/PYL800/blob/main/Codes/sharing.ipynb`


### Binder

Binder is a great way to share your code, provided that the github repository is *binderize*. Assuming it has been done, the link a the notebook is:
`https://mybinder.org/v2/gh/<user>/<repo-name>/<branch>?filepath=<file-path>`

For example, the link for this notebook is:
`https://mybinder.org/v2/gh/Hari31416/PYL800/HEAD?filepath=Codes/sharing.ipynb"`

**Note:** It can take some time to load up the binder. So, I'll suggest you use the above two methods.

## Printing the Notebook

You can use the export options to export the notebook as an html, pdf etc. However, you need to have to open the notebook in the browser first. Another way to export the notebook is to use the `nbconvert` command line tool.

Use the command `jupyter nbconvert --to webpdf <file-path>` to convert the notebook to a pdf file using web version. This way, the pdf produced is better rendered than usual export options.

**Note:** Sometime, the above command may give some error. In that case, add the flag `--allow-chromium-download`. The new command will be `jupyter nbconvert --allow-chromium-download --to webpdf <file-path>`.

You can use the `--to html` option to convert the notebook to html.

Using `--to latex` option, you can convert the notebook to latex. However, the latex file will not be that great!

Both in case of webpdf and html, you can pass `template` flag to specify the template to be used. The available templates are:

* `--template lab` (default)<br>
A full static HTML render of the notebook. This looks very similar to the jupyter lab interactive view.
<br>
* `--template classic`<br>
Simplified HTML, using the classic jupyter look and feel.
<br>
* `--template basic`<br>
Base HTML, rendering with minimal structure and styles.

### Doing Some Experiments

```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
plt.plot(x, y)
```

```python
%matplotlib widget
plt.plot(x, y)
```

```python
px.line(x = x, y = y, title='Sine Wave')
```

After exporting the pdf using all the three templates, I found that the default template is the best. Also, none of the above templates captures the last two interactive outputs. HTML also does not have the interactive outputs.
