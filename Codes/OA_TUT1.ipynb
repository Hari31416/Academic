```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rp(n1, n2, theta1, theta2):
    """
    Return the reflection coefficient of a plane wave
    """
    n = n1*np.cos(theta2) - n2*np.cos(theta1)
    d = n1*np.cos(theta2) + n2*np.cos(theta1)
    return n/d
```

```python
def tp(n1, n2, theta1, theta2):
    """
    Return the transmission coefficient of a plane wave
    """
    n = 2*n1*np.cos(theta1)
    d = n1*np.cos(theta2) + n2*np.cos(theta1)
    return np.abs(n/d)
```

```python
def rs(n1, n2, theta1, theta2):
    """
    Return the reflection coefficient of a plane wave
    """
    n = n1*np.cos(theta1) - n2*np.cos(theta2)
    d = n1*np.cos(theta1) + n2*np.cos(theta2)
    return n/d
```

```python
def ts(n1, n2, theta1, theta2):
    """
    Return the transmission coefficient of a plane wave
    """
    n = np.abs(2*n1*np.cos(theta1))
    d = np.abs(n1*np.cos(theta1) + n2*np.cos(theta2))
    return n/d
```

```python
theta = np.linspace(0, np.pi/2, 100)
n1=1
n2=1.5
theta2 = (theta*n1)/(n2)
rps = rp(n1, n2, theta, theta2)
tps = tp(n1, n2, theta, theta2)
rss = rs(n1, n2, theta, theta2)
tss = ts(n1, n2, theta, theta2)
```

```python
plt.figure(figsize=(8,6))
plt.plot(theta*180/np.pi, np.abs(rps), label='Reflection S')
plt.plot(theta*180/np.pi, np.abs(rss), label='Reflection P')
plt.plot(theta*180/np.pi, np.abs(tps), label='Transmission S')
plt.plot(theta*180/np.pi, np.abs(tss), label='Transmission P')
plt.legend()
plt.title('Reflection and Transmission Coefficients for Air to Glass')
plt.xlabel('Angle of Incident (degrees)')
plt.ylim(0,1);
```
